### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# Object Oriented Style:
#    Explicitly create objects (figures and axes) and call get/set methods on them 
#
# Traditional:
#    Call functions direct as in Matlab
# The recommended is the OO style
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# A Figure is the drawing canvas that may include one or more Axes-objects. Each Axes/subplot is built up by different Artists
#    - Axes: a plot area, may be an array using subplot functionality
#    - Axis: the scale with limits, ticks and labels
#    - Artists: object type for more or less everything in the Figure e.g axis, line, legend
```

### When plotting in Matplotlib, what is the expected input data type?

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4,4,50)
y = x**2

plt.figure()
plt.plot(x,y,'*')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of $x^2$')

fig_oo,ax_oo = plt.subplots()
ax_oo.plot(x,y,'r*')
ax_oo.set_xlabel('X')
ax_oo.set_ylabel('Y')
ax_oo.set_title('Plot of $x^2$ using OO style' )

```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, (ax1,ax2) = plt.subplots(2,1)
ax1.scatter(random_data_x,random_data_y,alpha=0.3, edgecolors='none')
ax2.bar(names,values,width=0.3)
```
